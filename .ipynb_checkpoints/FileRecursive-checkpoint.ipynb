```python
import os

def list_files_recursive(directory):
    file_list = []

    # Iterate over all items in the directory
    for item in os.listdir(directory):
        item_path = os.path.join(directory, item)

        # Check if the item is a file
        if os.path.isfile(item_path):
            file_list.append(item_path)
        # Check if the item is a directory
        elif os.path.isdir(item_path):
            # Recursively call the function for subdirectories
            subfolder_files = list_files_recursive(item_path)
            file_list.extend(subfolder_files)

            # Log the subfolder and number of files
            num_files_in_subfolder = len(subfolder_files)
            print(f"Subfolder: {item_path} | Number of Files: {num_files_in_subfolder}")
            print("---------------------------------")

    return file_list

def root_folder_files(directory):
    print(f"RootFolder: {directory} | Number of Files: {len([x for x in os.listdir(directory) if os.path.isfile(os.path.join(directory, x))])}")
    print("---------------------------------")

# Example: Replace 'your_directory_path' with the path of the folder you want to start the search from
directory_path = '/Users/habeebbabatunde/mlflow_web_app'
root_folder_files(directory_path)
all_files = list_files_recursive(directory_path)

# Display the list of files
#for file_path in all_files:
#    print(file_path)
```

```python
import os

def list_files_with_subfolders(directory_path):
    all_files = []
    
    counter = 0
    for root, dirs, files in os.walk(directory_path):
        file_paths = [os.path.join(root, x) for x in files]
        all_files.extend(file_paths)
        
        if counter==0:
            print(f"Rootfolder: {root} | Number of Files: {len(file_paths)}")
            print("-" * (len(f"Rootfolder: {root} | Number of Files: {len(file_paths)}") + 5))
        else:
            # Log the subfolder and number of files with dashes covering the text above
            print(f"Subfolder: {root} | Number of Files: {len(file_paths)}")
            print("-" * (len(f"Subfolder: {root} | Number of Files: {len(file_paths)}") + 5))
        counter+=1
    return all_files
```

```python
import numpy as np
from itertools import groupby

def are_arrays_equal(arr1, arr2):
    return np.array_equal(arr1, arr2)

def group_arrays_by_wavenumbers(arrays):
    # Enumerate to get both arrays and their indices
    enumerated_arrays = list(enumerate(arrays))

    # Sort the arrays based on their wavenumbers
    enumerated_arrays.sort(key=lambda x: tuple(x[1]))

    grouped_arrays = []
    for key, group in groupby(enumerated_arrays, lambda x: tuple(x[1])):
        indices, arrays_with_same_wavenumbers = zip(*group)
        grouped_arrays.append({'wavenumbers': key, 'indices': list(indices), 'arrays': list(arrays_with_same_wavenumbers)})

    return grouped_arrays

# Example: Replace the arrays with your actual data
array1 = np.array([11900, 4000, -8])
array2 = np.array([11980, 3900, -8])
array3 = np.array([11900, 4000, -8])
array4 = np.array([12000, 4100, -8])
array5 = np.array([11980, 3900, -8])
array6 = np.array([12100, 4200, -8])

arrays = [array1, array2, array3, array4, array5, array6]

# Group arrays with similar wavenumbers
grouped_arrays = group_arrays_by_wavenumbers(arrays)

# Display the groups
for idx, group in enumerate(grouped_arrays, 1):
    print(f"Group {idx}:")
    print(f"  Wavenumbers: {group['wavenumbers']}")
    print(f"  Indices: {group['indices']}")
    print(f"  Arrays: {group['arrays']}")
    print("---------------------------------")
```

```python
def are_all_arrays_equal(arrays):
    # Check if all arrays are equal to the first array
    return all(np.array_equal(arr, arrays[0]) for arr in arrays)

# Example: Replace the arrays with your actual data
array1 = np.array([11900, 4000, -8])
array2 = np.array([11900, 4000, -8]) #np.array([11980, 3900, -8])
array3 = np.array([11900, 4000, -8])

arrays_equal = are_all_arrays_equal([array1, array2, array3])

if arrays_equal:
    print("All arrays are equal.")
else:
    print("Arrays are not equal.")
```
